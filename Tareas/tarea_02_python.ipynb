**EJECUTA ESTE SCRIPT PARA GARANTIZAR LOS TESTS DE LAS PRUEBAS**

````python
# 🔁 Decorador simple para mostrar errores de tests de forma clara en Jupyter
def simple_test_runner(test_func):
    def wrapper():
        try:
            test_func()
            print("✅ Todos los tests pasaron")
        except (AssertionError, TypeError) as e:
            print(f"❌ Error durante los tests: {type(e).__name__}")
            raise e
    return wrapper
````

# Pregunta 1

Realizar un método (función) que revise si una cadena de carácteres es palíndromo.
Recuerda que un palíndromo es una palabra que al derecho o al revés es la misma. Ejemplo: kayak o taco cat.

Tip: no se deben tener en cuenta espacios o mayúsculas.

**Ejemplo:**

Entrada: kayak
Salida: True

Entrada: clase
Salida: False

Para poder testear, el método se debe llamar `IsPalindrome`.

Por ejemplo:

```python
def IsPalindrome():

    return

````python
# Escribe acá el método de manera correcta

def IsPalindrome(text):
   # Normalizar: quitar espacios y pasar todo a minúsculas
    text = text.replace(" ", "").lower()

    # Comparar con su reverso
    return text == text[::-1]

    return False  # claramente incorrecto
````

Una vez crees el método, ejecuta el script de abajo para verificar si todos los casos son correctos.

````python
import pytest

@simple_test_runner
def test_palindrome_checker():
    assert IsPalindrome("Kayak") == True
    assert IsPalindrome("Taco Cat") == True
    assert IsPalindrome("Prueba") == False
    assert IsPalindrome("A man a plan a canal Panama") == True
    assert IsPalindrome("No lemon, no melon") == True
    assert IsPalindrome("") == True  # Empty string case
    assert IsPalindrome("a") == True  # Single character case

test_palindrome_checker()
````

# Pregunta 2

**Valores únicos en al menos 2 de 3 listas**

Dadas tres listas de valores, encontrar todos los valores únicos que aparecen en al menos dos de las listas al mismo tiempo.

**Pista:**
Utiliza conjuntos para eliminar duplicados y realizar operaciones de intersección.
Encuentra la intersección de cada par de listas.
Une las intersecciones para obtener los valores únicos que aparecen en al menos dos de las listas.

````python
def unique_values(list1, list2, list3):
    #Convertir listas a conjuntos
    c1, c2, c3 = set(list1), set(list2), set(list3)

    #Intersección de conjuntos
    inter_12 = c1.intersection(c2)
    inter_13 = c1.intersection(c3)
    inter_23 = c2.intersection(c3)

    #Unión de intersecciones
    union = inter_12.union(inter_13, inter_23)

    #Convertir a lista
    return list(union)
````

Una vez crees el método, ejecuta el script de abajo para verificar si todos los casos son correctos.

````python
import pytest

@simple_test_runner
def test_unique_values():
    # Caso básico con valores comunes entre pares de listas
    assert sorted(unique_values([1, 2, 3], [2, 3, 4], [3, 4, 5])) == [2, 3, 4]

    # Caso sin valores comunes entre ninguna lista
    assert unique_values([1, 2], [3, 4], [5, 6]) == []

    # Caso donde todas las listas tienen los mismos elementos
    assert sorted(unique_values([1, 2], [1, 2], [1, 2])) == [1, 2]

    # Caso con elementos duplicados en las listas de entrada
    assert sorted(unique_values([1, 1, 2, 3], [2, 2, 3, 4], [3, 3, 4, 5])) == [2, 3, 4]

    # Caso con listas vacías
    assert unique_values([], [], []) == []

    # Caso donde una lista está vacía
    assert sorted(unique_values([1, 2], [], [2, 3])) == [2]

    # Caso con strings para probar otros tipos de datos (solo strings)
    assert sorted(unique_values(["a", "b"], ["b", "c"], ["c", "d"])) == ["b", "c"]

test_unique_values()
````

````python
#Aunque no se requiere por ser una prueba test, se realiza prueba
print(unique_values([1, 2, 3], [2, 3, 4], [3, 4, 5]))
print(unique_values([1, 2], [3, 4], [5, 6]))
print(unique_values([1, 2], [1, 2], [1, 2]))
print(unique_values([1, 1, 2, 3], [2, 2, 3, 4], [3, 3, 4, 5]))
print(unique_values([], [], []))
print(unique_values([1, 2], [], [2, 3]))
print(unique_values(["a", "b"], ["b", "c"], ["c", "d"]))
````
