**P8.** Crea dos arreglos de 5 elementos con `np.random.randint` (valores entre 1 y 20).  
Súmalos, réstalos y multiplícalos elemento por elemento. Imprime cada resultado.

```python
# Tu respuesta aquí
arreglo1 = np.random.randint(1, 20, 5)
print(arreglo1)
arreglo2 = np.random.randint(1, 20, 5)
print(arreglo2)

#suma
suma = arreglo1 + arreglo2
print("Suma:", suma)

#resta
resta = arreglo1 - arreglo2
print("Resta:", resta)

#multiplicación
multiplicacion = arreglo1 * arreglo2
print("Multiplicación:", multiplicacion)
```

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,  
usa **indexación booleana** para obtener solo las notas mayores o iguales a 6. Imprime el resultado.

```python
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

# Tu respuesta aquí
notas6 = notas[notas >= 6]
print("Notas mayores o iguales a 6:", notas6)
```
