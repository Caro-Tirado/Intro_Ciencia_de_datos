**P4.** Rellena los valores nulos de la columna `Age` con la **mediana** de esa columna.  
Verifica que ya no haya nulos en `Age`.

```python
# Tu respuesta aquí
df['Age'].fillna(df['Age'].median(), inplace=True)
df.isnull().sum()
```

## Parte 3 — Nuevas columnas y agrupaciones

**P5.** Crea una nueva columna llamada `FamilySize` que sea la suma de `SibSp` + `Parch` + 1 (el pasajero mismo).

```python
# Tu respuesta aquí
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

**P6.** Usando `groupby`, calcula la **tasa de supervivencia promedio** (columna `Survived`) por clase (`Pclass`).  
¿Qué clase tuvo mayor probabilidad de sobrevivir?

```python
# Tu respuesta aquí
df.groupby('Pclass')['Survived'].mean()
```

La clase que tuvo mayor probabilidad de sobrevivir fue la clase 1 con un promedio de supervivencia del 62,96%. La de menor probabilidad fue la clase 3 (promedio igual a 24.23%).

**P7.** Calcula la edad promedio y la tarifa (`Fare`) promedio, agrupando por `Sex` y `Pclass`.  
Usa `.groupby()` con `.agg()`.

```python
# Tu respuesta aquí
df.groupby(['Sex', 'Pclass']).agg({'Age': 'mean', 'Fare': 'mean'})
```

Para el sexo femenino se evidencia la edad promedio para clase 1 de 33 años, con una tarifa de 106.125798, la cual difiere en gran medida de la clase 1 del sexo masculino con una tarifa de 67.226127.
